## Do championship teams generally have a higher effective field goal percentage (eFG%) than the league average?

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define a dictionary to store championship team eFG% and league average eFG%
championship_efg = {}
league_avg_efg = {}

# Function to calculate championship team eFG% and league average eFG%
def calculate_efg(df):
    champion = df[df['Champion'] == 1]
    championship_efg[df['Year'].iloc[0]] = champion['eFG%'].values[0]
    league_avg_efg[df['Year'].iloc[0]] = df['eFG%'].mean()

# Iterate through the seasons
for year in range(1994, 2024):
    # Import the CSV file for the corresponding year as a pandas df
    df = pd.read_csv(f"./data/{year}.csv")
    
    # Call the function to calculate eFG% statistics
    calculate_efg(df)

# Calculate correlation
efg_championship = np.array(list(championship_efg.values()))
efg_league_avg = np.array(list(league_avg_efg.values()))
correlation = np.corrcoef(efg_championship, efg_league_avg)[0, 1]

# Plot graph with polynomial curve fit
plt.figure(figsize=(10, 6))
plt.scatter(years, efg_championship, color='blue', label='Championship Team eFG%')
plt.plot(years, efg_league_avg, color='red', linestyle='dashed', label='League Average eFG%')

# Fit a polynomial curve
z = np.polyfit(years, efg_championship, 2)
p = np.poly1d(z)
plt.plot(years, p(years), color='green', linestyle='dotted', label='Polynomial Curve Fit')

plt.xlabel('Year')
plt.ylabel('eFG%')
plt.title('Championship Team eFG% vs. League Average eFG% with Polynomial Curve Fit')
plt.legend()
plt.grid(True)
plt.show()

print("Correlation between championship team eFG% and league average eFG%:", correlation)
```

The correlation coefficient represents the strength and direction of the linear relationship between two variables. In this case, it quantifies the linear relationship between the eFG% of championship teams and the league average eFG% across the years. The correlation coefficient can take values between -1 and 1:

- If the correlation coefficient is close to 1, it indicates a strong positive linear relationship. In this context, it would mean that as the eFG% of championship teams increases, the league average eFG% also tends to increase, or vice versa.
- If the correlation coefficient is close to -1, it indicates a strong negative linear relationship. This would mean that as the eFG% of championship teams increases, the league average eFG% tends to decrease, or vice versa.
- If the correlation coefficient is close to 0, it indicates a weak linear relationship. In this case, there may not be a clear pattern of association between the two variables.

## Analysis: 

Correlation Coefficient:
- The correlation coefficient of 0.877 indicates a strong positive linear relationship between the eFG% of championship teams and the league average eFG% across the years.
- This suggests that there is a consistent trend where championship teams tend to have higher eFG% compared to the league average in NBA seasons from 1994 to 2023.

Graphical Representation:
- The plotted graph visually illustrates the relationship between championship team eFG% and league average eFG% over the years.
- The blue dots represent the eFG% of championship teams, while the red dashed line represents the league average eFG%.
- The green dotted line represents the polynomial curve fit, which appears to provide the best fit to the championship team eFG% data.

Interpretation:
- The upward trend in both championship team eFG% and league average eFG% suggests an overall improvement in shooting efficiency over the years.
- The fact that the championship team eFG% consistently exceeds the league average eFG% supports the notion that shooting efficiency plays a significant role in determining championship success in the NBA.

Implications:
- The strong positive correlation and the clear trend observed in the data emphasize the importance of effective field goal shooting in NBA championship contention.
- Teams aspiring for championship success may benefit from focusing on improving their shooting efficiency, as it appears to be a key factor distinguishing championship-winning teams from the rest of the league.

In conclusion, the analysis highlights the significance of shooting efficiency in NBA championship outcomes and provides valuable insights for teams and stakeholders interested in understanding the factors contributing to success in professional basketball.

## Are teams that are more dependent on 3-point shots less likely to win the championship?

```python
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

# Define a dictionary to store championship status and 3-point percentage dependence
championship_status = {}
three_point_dependence = {}

# Iterate through the years
for year in range(1994, 2024): 
    # Import the CSV file for the corresponding year as a pandas df 
    file_path = f"./data/{year}.csv"
    df = pd.read_csv(file_path) 
    
    # Calculate total points from 3-pointers and total points scored
    df['TotalPoints'] = df['3P'] * 3 + (df['PPG'] * 82 - df['3P']) * 2
    
    # Calculate 3-point percentage dependence (percentage of total points from 3-pointers)
    df['3PointDependence'] = df['3P'] * 3 / df['TotalPoints']
    
    # Identify championship team
    champion = df[df['Champion'] == 1]['Team'].iloc[0]
    
    # Store championship status and 3-point percentage dependence
    championship_status[year] = champion
    three_point_dependence[year] = df[df['Team'] == champion]['3PointDependence'].iloc[0]

# Convert dictionaries to DataFrame
df_dependence = pd.DataFrame(list(three_point_dependence.items()), columns=['Year', '3PointDependence'])

# Plot the scatter plot with a line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(df_dependence['Year'], df_dependence['3PointDependence'], color='blue', alpha=0.5)
plt.xlabel('Year')
plt.ylabel('3-Point Percentage Dependence (Percentage of Total Points)')
plt.title('3-Point Percentage Dependence Over Years')

# Fit a polynomial curve
z = np.polyfit(df_dependence['Year'], df_dependence['3PointDependence'], 1)
p = np.poly1d(z)
plt.plot(df_dependence['Year'], p(df_dependence['Year']), color='red', linestyle='dashed', label='Line of Best Fit')

plt.legend()
plt.grid(True)
plt.show()

# Calculate correlation
correlation = np.corrcoef(df_dependence['Year'], df_dependence['3PointDependence'])[0, 1]
print("Correlation between Year and 3-Point Percentage Dependence:", correlation)
```

Correlation Coefficient:
- The correlation coefficient of approximately 0.8458 indicates a strong positive linear relationship between the year and the 3-point percentage dependence of championship teams.
- This suggests that over the years from 1994 to 2023, NBA championship teams have shown an increasing trend in their dependence on 3-point shots as a percentage of their total points scored.
 
Graphical Representation:
- The plotted graph visually illustrates the relationship between the year and the 3-point percentage dependence of championship teams.
- The x-axis represents the year, while the y-axis represents the 3-point percentage dependence.
- The blue dots represent the 3-point percentage dependence of championship teams over the years, and the red dashed line represents the line of best fit.

Interpretation:
- The strong positive correlation coefficient and the upward trajectory of the line of best fit indicate a clear trend: championship-winning teams have increasingly relied on 3-point shots over the years.
- This trend reflects the evolving strategies and dynamics of the game, with teams recognizing the importance of 3-point shooting in achieving success in the NBA. 

Implications:
- The strong correlation between the year and 3-point percentage dependence suggests that modern NBA championship teams prioritize effective 3-point shooting as a key component of their offensive strategy.
- Teams aspiring for championship success should consider the strategic implications of this trend and prioritize player development, roster construction, and game planning accordingly.

In conclusion, the analysis underscores the growing significance of 3-point shooting in the NBA, with championship-winning teams demonstrating an increasing reliance on this aspect of the game over the past few decades.

## What role does player experience (measured by average team age) play in winning championships?

Correlation Coefficient:
- The correlation coefficient of approximately -0.889 indicates a strong negative linear relationship between the year and the average age of championship-winning teams.

- The x-axis represents the year, while the y-axis represents the average age of championship teams.
- The blue dots represent the average age of championship teams over the years, and the red dashed line represents the line of best fit.

Interpretation:
- The strong negative correlation coefficient and the downward trajectory of the line of best fit indicate a clear trend: championship-winning teams have become younger on average over the years.

## Is there a correlation between high team assists (AST) per game and winning the championship?

## What is the average age of championship teams, and how does this compare with non-champion teams?

```python
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Define a dictionary to store all the datasets 
season_statistics = {} 

# Iterate through the years
for year in range(1994, 2024): 
    
    # Import the CSV file for the corresponding year as a pandas df 
    df = pd.read_csv(f"./data/{year}.csv") 
    
    # Add the dataframe to our dictionary of dataframes 
    season_statistics[year] = df 

# Initialize variables to store age data
champion_ages = []
non_champion_ages = []
years = []

# Iterate through each year's data
for year, df in season_statistics.items():
    # Extract champion team and its age
    champion_team = df[df['Champion'] == 1]
    champion_age = champion_team['Age'].mean()
    if not np.isnan(champion_age):  # Check for NaN values
        champion_ages.append(champion_age)
    
    # Extract non-champion teams and their ages
    non_champion_teams = df[df['Champion'] == 0]
    non_champion_age = non_champion_teams['Age'].mean()
    if not np.isnan(non_champion_age):  # Check for NaN values
        non_champion_ages.append(non_champion_age)
    
    years.append(year)

# Calculate average age of championship and non-championship teams
avg_champion_age = sum(champion_ages) / len(champion_ages)
avg_non_champion_age = sum(non_champion_ages) / len(non_champion_ages)

# Fit a linear regression line to the data
slope_champion, intercept_champion, _, _, _ = linregress(years, champion_ages)
slope_non_champion, intercept_non_champion, _, _, _ = linregress(years, non_champion_ages)

# Calculate correlation scores
correlation_champion = np.corrcoef(years, champion_ages)[0, 1]
correlation_non_champion = np.corrcoef(years, non_champion_ages)[0, 1]

# Plot the data points and the line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(years, champion_ages, label='Championship Teams', color='blue')
plt.plot(years, intercept_champion + slope_champion * np.array(years), 'b--')
plt.scatter(years, non_champion_ages, label='Non-Championship Teams', color='red')
plt.plot(years, intercept_non_champion + slope_non_champion * np.array(years), 'r--')
plt.xlabel('Year')
plt.ylabel('Average Age')
plt.title('Average Age of Championship and Non-Championship Teams Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Print results
print("Average Age of Championship Teams:", avg_champion_age)
print("Average Age of Non-Championship Teams:", avg_non_champion_age)
print("Correlation between Year and Average Age (Championship Teams):", correlation_champion)
print("Correlation between Year and Average Age (Non-Championship Teams):", correlation_non_champion)
```

Average Age of Championship Teams:
- The average age of championship teams over the years is approximately 28.74 years.

Average Age of Non-Championship Teams:
- The average age of non-championship teams over the years is approximately 26.47 years.

Correlation between Year and Average Age (Championship Teams):
- The correlation coefficient between the year and the average age of championship teams is approximately -0.43. This negative correlation suggests a weak negative linear relationship between the year and the average age of championship teams. In other words, as the years progress, the average age of championship teams tends to slightly decrease.

Correlation between Year and Average Age (Non-Championship Teams):
- The correlation coefficient between the year and the average age of non-championship teams is approximately -0.85. This strong negative correlation suggests a significant negative linear relationship between the year and the average age of non-championship teams. In other words, as the years progress, the average age of non-championship teams tends to decrease considerably.

Summary:
- Average Age Trends: Both championship and non-championship teams have seen a decrease in average age over the years.
- Comparison: Championship teams tend to have a slightly higher average age compared to non-championship teams.
- Correlation Analysis: While there is a negative correlation between year and average age for both championship and non-championship teams, the correlation is stronger for non-championship teams, indicating a more pronounced decrease in average age over time.
- These findings suggest that age may play a role in team success, with younger teams possibly having an advantage in the NBA. However, other factors such as player skill, team chemistry, and coaching strategies also contribute to a team's performance and championship success.

## Are there any patterns in the strength of schedule (SOS) of championship-winning teams?

```python
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Define a dictionary to store all the datasets 
season_statistics = {} 

# Iterate through the years
for year in range(1994, 2024): 
    
    # Import the CSV file for the corresponding year as a pandas df 
    df = pd.read_csv(f"./data/{year}.csv") 
    
    # Add the dataframe to our dictionary of dataframes 
    season_statistics[year] = df 

# Initialize variables to store SOS data and years
champion_sos = []
years = []

# Iterate through each year's data
for year, df in season_statistics.items():
    # Extract champion team and its SOS
    champion_team = df[df['Champion'] == 1]
    sos_champion = champion_team['SOS'].mean()
    if not np.isnan(sos_champion):  # Check for NaN values
        champion_sos.append(sos_champion)
        years.append(year)

# Calculate average SOS of championship teams
avg_champion_sos = sum(champion_sos) / len(champion_sos)

# Fit a linear regression line to the data
slope_champion, intercept_champion, r_value, _, _ = linregress(years, champion_sos)

# Calculate correlation coefficient
correlation_champion = r_value ** 2  # Square of the Pearson correlation coefficient

# Plot the data points and the line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(years, champion_sos, label='Championship Teams', color='blue')
plt.plot(years, intercept_champion + slope_champion * np.array(years), 'b--')
plt.xlabel('Year')
plt.ylabel('Strength of Schedule (SOS)')
plt.title('Strength of Schedule (SOS) of Championship-Winning Teams Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Print results
print("Average SOS of Championship Teams:", avg_champion_sos)
print("Correlation between Year and SOS (Championship Teams):", correlation_champion)
```

Average SOS of Championship Teams:
- The average strength of schedule (SOS) of championship teams over the years is approximately -0.195.

Correlation between Year and SOS (Championship Teams):
- The correlation coefficient between the year and the strength of schedule (SOS) for championship-winning teams is approximately 0.004. This very low correlation suggests almost no linear relationship between the year and the SOS of championship teams.

Summary:
- Average SOS Trends: Championship teams have had a slightly negative average SOS over the years, indicating that they faced slightly weaker opponents on average.
- Correlation Analysis: The correlation coefficient near zero suggests that there is almost no linear relationship between the year and the SOS of championship teams. In other words, the SOS of championship teams does not appear to vary systematically with time.
- This analysis indicates that the strength of schedule may not be a significant factor in determining championship success in the NBA, at least based on the available data. Other factors such as team performance, player talent, and coaching strategies may play a more crucial role in achieving championship titles.
