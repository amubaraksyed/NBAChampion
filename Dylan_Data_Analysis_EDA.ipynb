```python
import os 
import pandas as pd 

# Define a dictionary to store all the datasets 
season_statistics = {} 

for year in range(1994, 2024): 
    
    # Import the CSV file for the corresponding year as a pandas df 
    df = pd.read_csv(f"./data/{year}.csv") 
    
    # Add the dataframe to our dictionary of dataframes 
    season_statistics[year] = df 
```

```python
from random import randint 

# Display a random season statistics to check if the code works 
season_statistics[randint(1994,2023)]
```

```python
# Are there any trends in the Pace or offensive style (e.g., higher FTr or 3PAr) among championship teams over multiple seasons?

import matplotlib.pyplot as plt

championship_teams_pace = {}
for year, df in season_statistics.items():
    championship_teams_pace[year] = df[df['Champion'] == 1]['Pace'].values[0]
    
championship_teams_ftr = {}
for year, df in season_statistics.items():
    championship_teams_ftr[year] = df[df['Champion'] == 1]['FTr'].values[0]

championship_teams_3par = {}
for year, df in season_statistics.items():
    championship_teams_3par[year] = df[df['Champion'] == 1]['3PAr'].values[0]
    
years = list(championship_teams_pace.keys())
paces = list(championship_teams_pace.values())

plt.plot(years, paces, marker='o', linestyle='-', color='blue')
plt.title('Pace of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Pace')
plt.grid(True)
plt.tight_layout()
plt.show()

years = list(championship_teams_ftr.keys())
paces = list(championship_teams_ftr.values())

plt.plot(years, paces, marker='o', linestyle='-', color='red')
plt.title('Free Throw Rate of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Free Throw Rate')
plt.grid(True)
plt.tight_layout()
plt.show()

years = list(championship_teams_3par.keys())
paces = list(championship_teams_3par.values())

plt.plot(years, paces, marker='o', linestyle='-', color='green')
plt.title('Three Point Attempt Rate of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Three Point Attempt Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```

While there are natural fluctations, the pace and three point attempt rate of championship winning teams has generally increased while the free throw rate has generally decreased.

```python
# What are the average statistics (e.g., PPG, DRtg, ORtg) of championship-winning teams compared to non-champions?

import matplotlib.pyplot as plt
import numpy as np

championship_teams_ppg = {}
for year, df in season_statistics.items():
    championship_teams_ppg[year] = df[df['Champion'] == 1]['PPG'].values[0]
    
non_championship_teams_ppg = {}
for year, df in season_statistics.items():
    non_championship_teams_ppg[year] = np.mean(df[df['Champion'] != 1]['PPG'].values)

championship_teams_ortg = {}
for year, df in season_statistics.items():
    championship_teams_ortg[year] = df[df['Champion'] == 1]['ORtg'].values[0]
    
non_championship_teams_ortg = {}
for year, df in season_statistics.items():
    non_championship_teams_ortg[year] = np.mean(df[df['Champion'] != 1]['ORtg'].values)
    
championship_teams_drtg = {}
for year, df in season_statistics.items():
    championship_teams_drtg[year] = df[df['Champion'] == 1]['DRtg'].values[0]
    
non_championship_teams_drtg = {}
for year, df in season_statistics.items():
    non_championship_teams_drtg[year] = np.mean(df[df['Champion'] != 1]['DRtg'].values)
    
championship_teams_nrtg = {}
for year, df in season_statistics.items():
    championship_teams_nrtg[year] = df[df['Champion'] == 1]['ORtg'].values[0] - df[df['Champion'] == 1]['DRtg'].values[0]
    
non_championship_teams_nrtg = {}
for year, df in season_statistics.items():
    non_championship_teams_nrtg[year] = np.mean(df[df['Champion'] != 1]['ORtg'].values) - np.mean(df[df['Champion'] != 1]['DRtg'].values)
    
years = list(championship_teams_ppg.keys())
ppg = list(championship_teams_ppg.values())
ppg2 = list(non_championship_teams_ppg.values())

plt.plot(years, ppg, marker='o', linestyle='-', color='green')
plt.plot(years, ppg2, marker='o', linestyle='-', color='red')
plt.title('Points Per Game of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Points Per Game')
plt.grid(True)
plt.tight_layout()
plt.show()

years = list(championship_teams_ortg.keys())
ortg = list(championship_teams_ortg.values())
ortg2 = list(non_championship_teams_ortg.values())

plt.plot(years, ortg, marker='o', linestyle='-', color='green')
plt.plot(years, ortg2, marker='o', linestyle='-', color='red')
plt.title('Offensive Rating of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Offensive Rating')
plt.grid(True)
plt.tight_layout()
plt.show()

years = list(championship_teams_drtg.keys())
drtg = list(championship_teams_drtg.values())
drtg2 = list(non_championship_teams_drtg.values())

plt.plot(years, drtg, marker='o', linestyle='-', color='green')
plt.plot(years, drtg2, marker='o', linestyle='-', color='red')
plt.title('Defensive Rating of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Defensive Rating')
plt.grid(True)
plt.tight_layout()
plt.show()

years = list(championship_teams_nrtg.keys())
nrtg = list(championship_teams_nrtg.values())
nrtg2 = list(non_championship_teams_nrtg.values())

plt.plot(years, nrtg, marker='o', linestyle='-', color='green')
plt.plot(years, nrtg2, marker='o', linestyle='-', color='red')
plt.title('Net Rating of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Net Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

While there are some outliers in championship winning teams' offensive or defensive rating being worse than league average, their net rating is always significantly above league average 0, which means that they have either a significantly positive offensive or defensive rating, if not both.

```python
# How does the rebounding performance (ORB%, DRB%) of championship teams compare with the league average?

import matplotlib.pyplot as plt
import numpy as np

championship_teams_orb = {}
for year, df in season_statistics.items():
    championship_teams_orb[year] = df[df['Champion'] == 1]['ORB%'].values[0]
    
non_championship_teams_orb = {}
for year, df in season_statistics.items():
    non_championship_teams_orb[year] = np.mean(df[df['Champion'] != 1]['ORB%'].values)
    
championship_teams_drb = {}
for year, df in season_statistics.items():
    championship_teams_drb[year] = df[df['Champion'] == 1]['DRB%'].values[0]
    
non_championship_teams_drb = {}
for year, df in season_statistics.items():
    non_championship_teams_drb[year] = np.mean(df[df['Champion'] != 1]['DRB%'].values)

years = list(championship_teams_orb.keys())
orb = list(championship_teams_orb.values())
orb2 = list(non_championship_teams_orb.values())

plt.plot(years, orb, marker='o', linestyle='-', color='green')
plt.plot(years, orb2, marker='o', linestyle='-', color='red')
plt.title('Offensive Rebound % of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Offensive Rebound %')
plt.grid(True)
plt.tight_layout()
plt.show()

years = list(championship_teams_drb.keys())
drb = list(championship_teams_drb.values())
drb2 = list(non_championship_teams_drb.values())

plt.plot(years, drb, marker='o', linestyle='-', color='green')
plt.plot(years, drb2, marker='o', linestyle='-', color='red')
plt.title('Defensive Rebound % of Championship-Winning Team Over Multiple Seasons')
plt.xlabel('Year')
plt.ylabel('Defensive Rebound %')
plt.grid(True)
plt.tight_layout()
plt.show()
```

While above-average offensive and defensive rebounds may contribute to a championship in some seasons, neither appear to be a requirement to winning a championship.
